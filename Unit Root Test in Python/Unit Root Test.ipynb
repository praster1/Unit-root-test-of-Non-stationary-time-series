```python
import rpy2
import rpy2.robjects as robjects
from rpy2.robjects import FloatVector
from rpy2.robjects.packages import importr

stats = importr('stats')
base = importr('base')
urca = importr('urca')
data = FloatVector([4.17,5.58,5.18,6.11,4.50,4.61,5.17,4.53,5.33,5.14])




A = urca.ur_df(data, lags=1, type='trend')
print(A)

# test stat
print("========= test statistics ==========")

print(A.slots['teststat'])
print(A.slots['teststat'].rx(1))
print(A.slots['teststat'].rx(2))
print(A.slots['teststat'].rx(3))

# critical value
print("========= critical value ==========")
cVal = A.slots['cval']
print(cVal)
print(cVal.rx(1,1))
print(cVal.rx(1,2))
print(cVal.rx(2,3))
print(cVal.rx(3,1))
```
